```python
import os
import pandas as pd
import numpy as np
```

Cek null yang ada di tiap kolom pada tiap file csv

```python
def cek_null_in_csv(directory='Laporan_keuangan'):
    file = {}

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            df = pd.read_csv(os.path.join(directory, filename))

            if df.isnull().values.any():
                file[filename] = df.isnull().sum()

    return file

file_null = cek_null_in_csv()
```

```python
for filename, value in file_null.items():
    print(f'File {filename} memiliki nilai null:')
    print(value)
    print()
```

Mengisi nilai null yang ada dengan nilai rata-rata kolom

```python
def fill_null_with_mean(directory='Laporan_keuangan'):
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            print(f'Filling null value in {filename} with mean value')
            df = pd.read_csv(os.path.join(directory, filename))
            
            df.replace('--', np.nan, inplace=True)
            columns_to_fill = df.columns.difference(['Date'])

            for column in columns_to_fill:
                if df[column].dtype == 'object':
                    df[column] = df[column].str.replace(',', '')
                    df[column] = pd.to_numeric(df[column], errors='coerce')

            df[columns_to_fill] = df[columns_to_fill].apply(lambda x: x.fillna(x.mean()) if x.dtype in [np.int64, np.float64] else x)

            df.to_csv(os.path.join(directory, filename), index=False)
            
fill_null_with_mean()
```

```python
file_null_after = cek_null_in_csv()

for filename, value in file_null_after.items():
    print(f'File {filename} setelah diisi nilai null:')
    print(value)
    print()
```
